```python
folders = [d.name for d in base_dir.iterdir() if d.is_dir()]
folders
```

```python
for folder in folders:
    try:
        last_number = re.search(r'\d+$', folder).group()
        # print(folder, last_number.group())
        
        for folder_bis in folders:
            if last_number in folder_bis and folder_bis != folder:
                print(f"    {last_number} from {folder} alreay present in {folder_bis}")
    except:
        print(f"No number in {folder} at the end")
        continue
```
